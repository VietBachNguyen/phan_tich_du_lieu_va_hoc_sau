```python
#2. Thông thường ta thường xử lý các vấn đề sau về dữ liệu
#1. Thiếu dòng tiêu đề ở file csv
df = pd.read_csv('patient_heart_rate.csv',header=None,on_bad_lines='skip')
df.columns = ['id', 'name', 'age', 'weight','hr1', 'hr2', 'hr3','hr4', 'hr5', 'hr6']

```

```python
 #2. Nhiều biến lưu ở một cột
df[['weight_value', 'weight_unit']] = df['weight'].str.extract(
    r'([\d\.]+)\s*([a-zA-Z]+)'
)
df
```

```python
df['weight_value'] = pd.to_numeric(df['weight_value'], errors='coerce')
df['weight_unit'] = df['weight_unit'].str.lower()
df['weight_kg'] = df['weight_value']
df.loc[df['weight_unit'].str.contains('lb', na=False), 'weight_kg'] *= 0.453592
df


```

```python
#4. Dữ liệu có một dòng trống
df.replace('-', pd.NA, inplace=True)
df.dropna(how='all', inplace=True)
df
```

```python
#5. Dữ liệu trùng lập
# Xóa các dòng trùng lặp
df.drop_duplicates(inplace=True)
df
```

```python
#6Các ký tự không phải ASCII
df['name'] = (df['name'].astype(str).str.encode('ascii', errors='ignore') .str.decode('ascii'))
df
```

```python
# 7.Giá trị bị mất
df.fillna(0, inplace=True)
df
```

```python
#8. Tiêu đề cột là giá trị chứ không phải tên biến
df.rename(columns={'m0006': 'Male_0_6','m0612': 'Male_6_12','m1218': 'Male_12_18','f006': 'Female_0_6','f0612': 'Female_6_12','f1218': 'Female_12_18'}, inplace=True)
df
```

```python
#3. Vấn đề 1: Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề “Missing header in the csv file”
import pandas as pd

column_names = ['Id','Name','Age','Weight','m0006','m0612','m1218','f0006','f0612','f1218']

df = pd.read_csv(
    "patient_heart_rate.csv",
    names=column_names,
    engine="python",
    on_bad_lines="warn"   # or "skip"
)

df.head()
```

```python

#4.Vấn đề 2: Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao gồm “Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột
```

```python
df[['Firstname','Lastname']] = df['Name'].astype(str).str.split(expand=True, n=1)
df.drop(columns=['Name'], inplace=True)

print(df.head())
```

```python
#5.Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn “kg”
```

```python
for i in range(len(df)):
    x = str(df.loc[i, 'Weight'])
    if x.endswith('lbs'):
        x = x.replace('lbs', '')
        x = float(x)
        y = int(x / 2.2)
        df.loc[i, 'Weight'] = str(y) + 'kgs'
print(df)
```

```python
#6.Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp cóthể đưa ra là xóa bỏ
```

```python
df.dropna(how='all', inplace=True)
print(df)
```

```python
#7.Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname,age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phảidựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.
```

```python
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(df)
```

```python
#8. Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII.Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)
```

```python
# Vấn đề 6: Loại bỏ ký tự không thuộc ASCII

df['Firstname'] = (df['Firstname'].astype(str).str.encode('ascii', errors='ignore').str.decode('ascii'))
df['Lastname'] = (df['Lastname'].astype(str).str.encode('ascii', errors='ignore').str.decode('ascii'))
print(df)
```
