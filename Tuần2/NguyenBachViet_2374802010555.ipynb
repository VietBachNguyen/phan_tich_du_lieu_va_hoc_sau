```python
def Q1(g):
  return np.percentile(g, 25)
```

```python
def Q2(g):
  return np.percentile(g, 50)
```

```python
def Q3(g):
  return np.percentile(g, 75)
```

```python
def std(g):
  return np.std(g)
```

```python
pd.pivot_table(data=dftuyensinh,index='KT', values = 'DH1',
 aggfunc=['sum', 'mean', 'median', 'min', 'max', std, Q1, Q2, Q3 ])
```

```python
#4.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT và KV
```

```python
pd.pivot_table(data=dftuyensinh,index=['KT','KV'], values='DH1',
               aggfunc=['sum', 'mean','median','min','max', std, Q1, Q2, Q3])
```

```python
#Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT, KV và DT
```
